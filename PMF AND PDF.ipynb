Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

Probability Mass Function (PMF) and Probability Density Function (PDF) are both functions that describe the probability distribution of a random variable.

PMF is used to describe the probability distribution of a discrete random variable. It assigns a probability to each possible outcome of the random variable. The sum of probabilities for all possible outcomes is equal to 1.

For example, let's consider a fair six-sided die. The PMF of the random variable representing the number rolled on the die can be represented as follows:

Outcome	PMF
1	1/6
2	1/6
3	1/6
4	1/6
5	1/6
6	1/6
Here, the PMF assigns an equal probability of 1/6 to each of the possible outcomes of rolling a die.

On the other hand, PDF is used to describe the probability distribution of a continuous random variable. It represents the relative likelihood of a continuous random variable taking on a particular value. Unlike PMF, PDF doesn't assign probabilities to specific values, but instead provides the probability density at each point on the continuous scale. The total area under the PDF curve is equal to 1.

For example, let's consider the height of adult males in a population. The PDF of this continuous random variable can be represented as follows:

PDF of Male Heights

In this example, the PDF shows that the probability of an adult male being between 170 and 180 cm tall is higher than the probability of him being between 160 and 170 cm tall. However, the PDF doesn't provide probabilities for specific heights, but rather the relative likelihood of a man being within a given range of heights.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for male heights
data = np.random.normal(175, 7, 1000)

# Create a histogram to visualize the data
n, bins, patches = plt.hist(data, bins=30, density=True, alpha=0.5)

# Fit a normal distribution to the data
mu, sigma = np.mean(data), np.std(data)
pdf = 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-(bins - mu)**2 / (2*sigma**2))

# Plot the PDF curve
plt.plot(bins, pdf, color='red', linewidth=2)

# Set the title and labels for the plot
plt.title('PDF of Male Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')

# Show the plot
plt.show()
```

The Bernoulli distribution is a discrete probability distribution that models a single experiment with two possible outcomes, usually labeled as success and failure. The probability of success is denoted by p, and the probability of failure is 1-p. The Bernoulli distribution is named after the Swiss mathematician Jacob Bernoulli, who introduced it in his work on probability theory.

An example of a Bernoulli distribution is flipping a coin, where success can be defined as getting heads and failure as getting tails. If the coin is fair, the probability of success (getting heads) is 0.5, and the probability of failure (getting tails) is also 0.5.

The Bernoulli distribution is closely related to the binomial distribution, which models the number of successes in a series of independent Bernoulli trials. The main difference between the two distributions is that the Bernoulli distribution models a single trial, while the binomial distribution models a series of n independent trials, each with the same probability of success, denoted by p.

The binomial distribution has two parameters: n, the number of trials, and p, the probability of success in each trial. The probability of getting k successes in n trials is given by the binomial distribution formula:

P(k) = (n choose k) p^k (1-p)^(n-k)

where (n choose k) is the binomial coefficient, which represents the number of ways to choose k successes out of n trials.

An example of a binomial distribution is the number of heads obtained in 10 flips of a fair coin. Here, n = 10, and p = 0.5. We can use the binomial distribution formula to calculate the probability of getting k heads in 10 flips, for different values of k.

In summary, the Bernoulli distribution models a single experiment with two possible outcomes, while the binomial distribution models the number of successes in a series of independent Bernoulli trials with the same probability of success. The binomial distribution is a generalization of the Bernoulli distribution to multiple trials.

z = (x - μ) / σ

The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that the sampling distribution of the mean of a large number of independent and identically distributed (iid) random variables will be approximately normal, regardless of the underlying distribution of the population.

In other words, the CLT states that as the sample size increases, the distribution of sample means becomes increasingly normal, with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size. This holds true for many different types of underlying distributions, as long as the sample size is large enough.

The significance of the Central Limit Theorem lies in its practical applications. It allows us to make inferences about the population mean based on the sample mean, even if the underlying distribution is not known or is not normal. It is widely used in statistical hypothesis testing, confidence intervals, and estimation.

For example, if we want to estimate the mean weight of all people in a certain country, we can take a random sample of individuals and calculate their mean weight. Even if the distribution of weights in the population is not known, we can use the CLT to approximate the sampling distribution of the mean weight and make inferences about the population mean with a certain degree of confidence.

In summary, the Central Limit Theorem is a fundamental theorem in statistics that states that the sampling distribution of the mean of a large number of iid random variables will be approximately normal, regardless of the underlying distribution of the population. Its significance lies in its practical applications, allowing us to make inferences about the population mean based on the sample mean, even if the underlying distribution is not known or is not normal.
